```python
import requests

# API endpoint URL
url = "http://127.0.0.1:5000/search"

params = {
    "q": "cerita+sukses+umkm",
    "detail": "true",
    "limit": 10
}

try:
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        # Print the JSON response data
        data = response.json()
        print("Status:", data["status"])
        print("Number of Results:", data["length"])
        # for item in data["data"]:
        #     print("Title:", item["judul"])
        #     # print("URL:", item.get("url", "No URL available"))
        #     print(item['body'])
        #     print()
    else:
        print("Error:", response.status_code, response.json().get("error", "Unknown error"))

except requests.exceptions.RequestException as e:
    print("Request failed:", e)
```

```python
toskip = [
    'SCROLL', 'TO', 'CONTINUE', 'WITH', 'CONTENT', 'ADVERTISEMENT', '\\', ''
]
toskip
```

```python
z=[]
FINAL_DATA = []
for i,j in enumerate(data['data']):
    print(i)
    o = j['body'].split("\n")
    o = ' '.join(o)
    o = o.split("\r")
    o = ' '.join(o)
    o = o.split(" ")
    # print(o)
    a = [i for i in o if i not in toskip]
    z = ' '.join(a)
    # print(z)
    FINAL_DATA.append(z)
data_dict = {"data" : FINAL_DATA}
# data_dict
```

```python
import json
# write FINAL_DATA to json file
filename = f'raw_dataset/news_result_{params['q']}.json'
with open(filename, "w") as final:
	json.dump(data_dict, final)
```
